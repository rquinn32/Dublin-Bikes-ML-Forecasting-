## 1. Data Preparation/Cleaning

### 1.1 Introduction - import libraries and read in CSV file

```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def load_data(file):
    # Read CSV file into a pandas dataframe
    # Print first few rows

    try:
        # If the csv file has any special symbols, it helps to specify the file encoding.
        df = pd.read_csv(file, 
            encoding="utf-8",
            delimiter=",",
            skipinitialspace=True
        )
        
        print(f"✓ Successfully loaded dataset with {df.shape[0]} rows and {df.shape[1]} columns")

        return df
    
    except FileNotFoundError:
        print("ERROR: The file '" + file + "' was not found.")
        print("Please ensure the file is in the current working directory.")
    except Exception as e:
        print(f"ERROR: An unexpected error occurred: {e}")

df_test = load_data("final_merged_data.csv")
df_raw = df_test.copy()
```

### 1.2 Data Validation


```python
def inspect_data(df):
    # Data validation checks

    print("Dataset Validation Checks:")
    print("="*60)

    # Check 1: Dataset is not empty
    if df.empty:
        print("⚠ WARNING: Dataset is empty!")
    else:
        print(f"✓ Dataset contains {df.shape[0]:,} rows")

    # Check 2: Check for duplicate rows
    duplicate_count = df.duplicated().sum()
    if duplicate_count > 0:
        print(f"⚠ WARNING: Found {duplicate_count} duplicate rows")
    else:
        print("✓ No duplicate rows found")

    # Check 3: Check if all columns are unnamed
    unnamed_cols = [col for col in df.columns if 'Unnamed' in str(col)]
    if unnamed_cols:
        print(f"⚠ WARNING: Found {len(unnamed_cols)} unnamed columns: {unnamed_cols}")
    else:
        print("✓ All columns have names")

    # Check 4: Check for missing data
    missing_count = df.isna().sum().sum()
    if missing_count > 0:
        print(f"⚠ WARNING: Found {missing_count} missing values")
    else:
        print("✓ No missing values found")

    # Check 5: Check for constant columns (cardinality of 1)
    constant_cols = df.columns[df.nunique() == 1].tolist()

    if constant_cols:
        print(f"\n⚠ WARNING: Constant columns found: {constant_cols}")
    else:
        print(f"✓ No constant columns found — all {df.shape[1]} features contain variation")

    # Check 5: Basic column overview
    print(f"\n✓ Dataset shape: {df.shape}")
    print(f"✓ Number of rows (instances): {df.shape[0]}")
    print(f"✓ Number of columns (features): {df.shape[1]}")

inspect_data(df_test)
```

### 1.3 Data Base Cleaning

```python
def base_clean(df):
    df = df.copy()
    df.columns = df.columns.str.strip().str.lower()

    # Parse last_reported into a datetime column
    df["date_time"] = pd.to_datetime(df["last_reported"], errors="coerce")

    # Drop duplicates and missing key values (not present in our dataset but included in case of use with future data)
    df = df.dropna(subset=["date_time", "station_id", "num_bikes_available", "capacity"])
    df = df.drop_duplicates()

    # Drop constant columns
    constant_cols = [c for c in df.columns if df[c].nunique() == 1]
    df = df.drop(columns=constant_cols)

    # Drop invalid columns
    df = df[(df['num_bikes_available'] >= 0) & 
        (df['num_bikes_available'] <= df['capacity'])]

    #Change station ID feature to categorical
    df["station_id"] = df["station_id"].astype("category")
    
    return df

df_test = base_clean(df_test)
```

### 1.4 Feature Investigation

```python
# Investigate correlation of features with target
corr = df_test.corr(numeric_only=True)["num_bikes_available"].sort_values()
print(corr)
```

```python
# Visualisation of feature pairs
corr_matrix = df_test.corr(numeric_only=True)

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Correlation Matrix")
plt.savefig('CorrelationMatrix.pdf', dpi=150, bbox_inches='tight')
plt.show()
```

A correlation matrix was generated to examine relationships between features and identify potential redundancy or risk of data leakage.
- A strong inverse correlation is noted between num_bikes_available and num_docks_available, identified as a potential cause of data leakage
- Soil, earth and grass temperatures exhibit negligible correlation with target feature
- High degree of multicollinearity among environmental variables, in particular statistical measurements of temperature, humidity and barometric pressure - highlights redundancy
- Geographical features (lat, lon) and capacity show moderate correlation with num_bikes_available
- Temporal features (hour, day) and environmental variables show low linear correlation with the target, however correlation does not capture non-linear relationships or time based patterns

### 1.4 Feature preparation

Cleaning decisions made:
- Parse last_reported into datetime column (possibly to be used to derive lag features during modelling)
- Drop rows with missing values or invalid values (e.g. num_bikes_available > capacity)
- Drop duplicate rows
- Drop constant columns
- Drop num_docks_available to prevent data leakage
- Drop redundant features (e.g. std columns)
- Drop categorical features not relevent for modelling (e.g. name, address, last reported)
- Drop soil/earth/grass features as not thought to be relevent for modelling

```python
def drop_columns(df):

    #Drop available docks feature - avoid data leakage
    df = df.drop(columns=["num_docks_available"], errors="ignore")
    
    # Drop quality indicator columns (should already be removed when removing constants)
    quality_indicator_cols = [c for c in df.columns if "quality_indicator" in c.lower()]
    df = df.drop(columns=quality_indicator_cols)

    # Drop earth/soil feature columns  - not statistically relevant 
    earth_cols = [c for c in df.columns if any(x in c.lower() for x in ["soil", "earth", "grass"])]
    df = df.drop(columns=earth_cols)

    # Drop std columns - redundancy
    std_cols = [c for c in df.columns if "std_deviation" in c.lower()]
    df = df.drop(columns=std_cols)

    # Drop other irrelevant columns
    df = df.drop(columns=["name", "address", "last_reported"], errors="ignore")

    return df
```

### 1.5 Derived features

```python
# Week day/ weekend indicators
def add_time_features(df):

    df["day_of_week"] = df["date_time"].dt.dayofweek
    df["is_weekend"] = (df["day_of_week"] >= 5).astype(int)

    df["hour_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
    df["hour_cos"] = np.cos(2 * np.pi * df["hour"] / 24)

    return df

# Average environmental features
def add_average_features(df):

    #Average temp
    df["avg_temp"] = (df["max_air_temperature_celsius"] + df["min_air_temperature_celsius"]) / 2

    #Average humidity
    df["avg_humidity"] = (df["max_relative_humidity_percent"] + df["min_relative_humidity_percent"]) / 2

    #Average pressure
    df["avg_barometric_pressure"] = (df["max_barometric_pressure_hpa"] + df["min_barometric_pressure_hpa"]) / 2

    # Drop min/max/std features
    df = df.drop(columns=[
        "max_air_temperature_celsius",
        "min_air_temperature_celsius",
        "max_relative_humidity_percent",
        "min_relative_humidity_percent",
        "max_barometric_pressure_hpa", 
        "min_barometric_pressure_hpa"
        ])
    return df
```

### 1.6 Apply changes and save new file

```python
df_cleaned = (
    load_data("final_merged_data.csv")
    .pipe(base_clean)
    .pipe(drop_columns)
    .pipe(add_time_features)
    .pipe(add_average_features)
)
```

```python
print("="*60)
print(f"Before cleaning: {df_raw.shape[0]:,} rows, {df_raw.shape[1]} columns")

rows_removed = df_raw.shape[0] - df_cleaned.shape[0]
cols_removed = df_raw.shape[1] - df_cleaned.shape[1]

print("="*60)
print(f"Rows removed: {rows_removed}")
print(f"Columns removed: {cols_removed}")

print("="*60)
print(f"After cleaning:  {df_cleaned.shape[0]:,} rows, {df_cleaned.shape[1]} columns")

print("="*60)
print(df_cleaned.info())
```

```python
#Save new cleaned file 
df_cleaned.to_csv("CleanedData.csv", index=False)
```

## 2. Model Development

```python
# Load dataset
df = pd.read_csv("CleanedData.csv")
df["date_time"] = pd.to_datetime(df["date_time"])
df = df.sort_values(['date_time'])
```

#### 2.1 Long Term Prediction Model

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Load dataset
df = pd.read_csv("CleanedData.csv")
df["date_time"] = pd.to_datetime(df["date_time"])
df = df.sort_values(['date_time'])

# Features
features = [
    "station_id",
    "capacity",
    "hour_sin",
    "hour_cos",
    "minute",
    "day_of_week",
    "is_weekend",
    "avg_temp",
    "avg_humidity",
    "avg_barometric_pressure"
]

X = df[features]
y = df['num_bikes_available']

# One-hot encode categorical features
categorical_cols = ['station_id', 'day_of_week']
X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# 80/20 chronological split
split_index = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

# Models 
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(
        n_estimators=200,
        max_depth=None,
        min_samples_split=5,
        random_state=42,
        n_jobs=-1
        ),
    "Decision Tree": DecisionTreeRegressor(
        max_depth=10,
        min_samples_split=10,
        random_state=42), 
    "Ridge Regression": Pipeline([
        ("scaler", StandardScaler()),
        ("ridge", Ridge(alpha=1.0))
        ])
}

# Train models
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    
    results[name] = {
        "MAE": mean_absolute_error(y_test, preds),
        "RMSE": np.sqrt(mean_squared_error(y_test, preds)),
        "R2": r2_score(y_test, preds)
    }

# Print metrics
for name, metrics in results.items():
    print(f"\n{name}")
    for k, v in metrics.items():
        print(f"{k}: {v:.4f}")

# Metrics table
results_df = pd.DataFrame(results).T.reset_index().rename(columns={"index": "Model"})
results_df = results_df.sort_values("R2", ascending=False)
display(results_df.round(3))

# Coefficients / importances
for name, model in models.items():
    print(f"\n{name}")

    if name == "Ridge Regression":
        vals = model.named_steps["ridge"].coef_
        df_out = pd.DataFrame({
            "Feature": X_train.columns,
            "Coefficient": vals
        })
        df_out["Abs"] = df_out["Coefficient"].abs()
        display(df_out.sort_values("Abs", ascending=False).drop(columns="Abs").head(10).round(3))

    elif hasattr(model, "coef_"):
        df_out = pd.DataFrame({
            "Feature": X_train.columns,
            "Coefficient": model.coef_
        })
        df_out["Abs"] = df_out["Coefficient"].abs()
        display(df_out.sort_values("Abs", ascending=False).drop(columns="Abs").head(10).round(3))

    elif hasattr(model, "feature_importances_"):
        df_out = pd.DataFrame({
            "Feature": X_train.columns,
            "Importance": model.feature_importances_
        })
        display(df_out.sort_values("Importance", ascending=False).head(10).round(3))
```

#### 2.2 Long Term Prediction Model Evaluation

The long-term prediction models achieved moderate performance, with linear and ridge regression performing best (R^2 ~ 0.30), indicating that bike availability is mainly influenced by stable factors such as station location, capacity, and time of day. Tree-based models performed worse, indicating that complex nonlinear relationships are not very important in this case. Feature importance showed that station ID, capacity, and hour are the most important predictors, while weather has a smaller effect. Overall, the models predict general patterns in bike usage, but the relatively low R² shows that bike availability is harder to predict without recent data.

#### 2.3 Short term prediction model

```python
# Sort by station and time
df = df.sort_values(["station_id", "date_time"]).reset_index(drop=True)

# Create lag features
df["bikes_avail_10mins_ago"] = df.groupby("station_id")["num_bikes_available"].shift(1)
df["bikes_avail_30mins_ago"] = df.groupby("station_id")["num_bikes_available"].shift(3)
df["bikes_avail_1h_ago"] = df.groupby("station_id")["num_bikes_available"].shift(6)
df["bikes_avail_24h_ago"] = df.groupby("station_id")["num_bikes_available"].shift(144)

# Create future targets
df["target_10min"] = df.groupby("station_id")["num_bikes_available"].shift(-1)
df["target_30min"] = df.groupby("station_id")["num_bikes_available"].shift(-3)
df["target_60min"] = df.groupby("station_id")["num_bikes_available"].shift(-6)

lag_cols = [
    "bikes_avail_10mins_ago",
    "bikes_avail_30mins_ago",
    "bikes_avail_1h_ago",
    "bikes_avail_24h_ago"
]

target_cols = ["target_10min", "target_30min", "target_60min"]

# Drop rows with missing lags or targets
df = df.dropna(subset=lag_cols + target_cols).reset_index(drop=True)

# Sort by time for split
df = df.sort_values("date_time").reset_index(drop=True)

# Time-based split
cutoff_time = df["date_time"].quantile(0.8)
train_df = df[df["date_time"] <= cutoff_time].copy()
test_df = df[df["date_time"] > cutoff_time].copy()

# Features for short-term prediction
features = [
    "station_id",
    "capacity",
    "hour_sin",
    "hour_cos",
    "day_of_week",
    "is_weekend",
    "avg_temp",
    "avg_humidity",
    "avg_barometric_pressure",
    "bikes_avail_10mins_ago",
    "bikes_avail_30mins_ago",
    "bikes_avail_1h_ago",
    "bikes_avail_24h_ago"
]

# Build X from split dataframes
X_train = train_df[features].copy()
X_test = test_df[features].copy()

# One-hot encode
X_train = pd.get_dummies(X_train, columns=["station_id", "day_of_week"], drop_first=True)
X_test = pd.get_dummies(X_test, columns=["station_id", "day_of_week"], drop_first=True)

# Align columns
X_train, X_test = X_train.align(X_test, join="left", axis=1, fill_value=0)
```

```python
# Define function to easily iterate through target times
def run_model(target_col, horizon_label):
    y_train = train_df[target_col].copy()
    y_test = test_df[target_col].copy()

    rows = []

    for name, model in models.items():
        model.fit(X_train, y_train)
        preds = model.predict(X_test)

        rows.append({
            "Horizon": horizon_label,
            "Model": name,
            "MAE": mean_absolute_error(y_test, preds),
            "RMSE": np.sqrt(mean_squared_error(y_test, preds)),
            "R2": r2_score(y_test, preds)
        })

    return pd.DataFrame(rows)
```

```python
# Test horizons
results_10 = run_model("target_10min", "10 Minutes")
results_30 = run_model("target_30min", "30 Minutes")
results_60 = run_model("target_60min", "60 Minutes")

final_results = pd.concat([results_10, results_30, results_60], ignore_index=True)

# Optional sorting
final_results = final_results.sort_values(by=["Horizon", "MAE"]).reset_index(drop=True)

display(final_results.round(3))
```

#### 2.4 Short Term Prediction Model Evaluation


The short-term prediction models performed very well across all time horizons, showing that recent bike availability is very important for making accurate predictions. At the 10-minute horizon, all models achieved very high accuracy (R² ≈ 0.97, MAE ≈ 0.7), meaning bike availability is highly predictable in the short term. As the horizon increased to 30 and 60 minutes, performance decreased slightly (R² ≈ 0.89–0.95), which is expected as predictions become more uncertain over time. Linear and ridge regression performed as well as or better than the more complex tree-based models, suggesting that short-term changes are mostly simple and follow clear linear patterns. These results show that bike availability depends strongly on recent data, making short-term predictions much more reliable compared to longer-term ones.

#### 2.5 Time Series Prediction Model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import statsmodels.api as sm

#Test one station
station_id_example = df['station_id'].unique()[0] 

df_station = df[df['station_id'] == station_id_example].set_index('date_time')
df_station = df_station.asfreq('10min')

# Fill missing values
y_station = df_station['num_bikes_available'].interpolate()

# Split into train and test
train_size = int(len(y_station) * 0.8)
train, test = y_station[:train_size], y_station[train_size:]

# Fit model
model = sm.tsa.ARIMA(train, order=(1,0,1))
result = model.fit()

# Forecast
forecast = result.forecast(steps=len(test))

print(result.summary())
print("MAE:", mean_absolute_error(test, forecast))
print("Next 10 forecast values:\n", forecast.iloc[:10])
```

The ARIMA model relies heavily on past values (≈0.99), but performs worse (MAE ≈ 5.9) than machine learning models. This is because it does not use external factors like time or weather. Although it produces smooth predictions, it does not respond well with sudden changes, making it less effective than our other models.

#### 2.6 Logistic Regression Model 

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.linear_model import LogisticRegression

# Sort by station and time
df = df.sort_values(["station_id", "date_time"]).reset_index(drop=True)

# Create lag features
df["bikes_avail_10mins_ago"] = df.groupby("station_id")["num_bikes_available"].shift(1)
df["bikes_avail_30mins_ago"] = df.groupby("station_id")["num_bikes_available"].shift(3)
df["bikes_avail_1h_ago"] = df.groupby("station_id")["num_bikes_available"].shift(6)
df["bikes_avail_24h_ago"] = df.groupby("station_id")["num_bikes_available"].shift(144)

# Create future numeric targets
df["target_10min"] = df.groupby("station_id")["num_bikes_available"].shift(-1)
df["target_30min"] = df.groupby("station_id")["num_bikes_available"].shift(-3)
df["target_60min"] = df.groupby("station_id")["num_bikes_available"].shift(-6)

# Convert to binary classification targets (1 = bikes available, 0 = no bikes available)
df["target_10min_cls"] = (df["target_10min"] > 0).astype(int)
df["target_30min_cls"] = (df["target_30min"] > 0).astype(int)
df["target_60min_cls"] = (df["target_60min"] > 0).astype(int)

lag_cols = [
    "bikes_avail_10mins_ago",
    "bikes_avail_30mins_ago",
    "bikes_avail_1h_ago",
    "bikes_avail_24h_ago"
]

target_cols = ["target_10min", "target_30min", "target_60min"]

# Drop rows with missing lags or targets
df = df.dropna(subset=lag_cols + target_cols).reset_index(drop=True)

# Sort by time for split
df = df.sort_values("date_time").reset_index(drop=True)

# Time-based split
cutoff_time = df["date_time"].quantile(0.8)
train_df = df[df["date_time"] <= cutoff_time].copy()
test_df = df[df["date_time"] > cutoff_time].copy()

# Features
features = [
    "station_id",
    "capacity",
    "hour_sin",
    "hour_cos",
    "day_of_week",
    "is_weekend",
    "avg_temp",
    "avg_humidity",
    "avg_barometric_pressure",
    "bikes_avail_10mins_ago",
    "bikes_avail_30mins_ago",
    "bikes_avail_1h_ago",
    "bikes_avail_24h_ago"
]

# Build X
X_train = train_df[features].copy()
X_test = test_df[features].copy()

# One-hot encode categorical columns
X_train = pd.get_dummies(X_train, columns=["station_id", "day_of_week"], drop_first=True)
X_test = pd.get_dummies(X_test, columns=["station_id", "day_of_week"], drop_first=True)

# Align columns
X_train, X_test = X_train.align(X_test, join="left", axis=1, fill_value=0)

# Logistic Regression model
log_model = Pipeline([
    ("scaler", StandardScaler()),
    ("logreg", LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42))
])

# Function to run logistic regression for each horizon
def run_logistic_model(target_col, horizon_label):
    y_train = train_df[target_col].copy()
    y_test = test_df[target_col].copy()

    log_model.fit(X_train, y_train)
    preds = log_model.predict(X_test)
    probs = log_model.predict_proba(X_test)[:, 1]

    results = pd.DataFrame([{
        "Horizon": horizon_label,
        "Model": "Logistic Regression",
        "Accuracy": accuracy_score(y_test, preds),
        "Precision": precision_score(y_test, preds, zero_division=0),
        "Recall": recall_score(y_test, preds, zero_division=0),
        "F1": f1_score(y_test, preds, zero_division=0)
    }])

    return results, preds, probs, y_test

# Run for each horizon
results_10, preds_10, probs_10, y_test_10 = run_logistic_model("target_10min_cls", "10 Minutes")
results_30, preds_30, probs_30, y_test_30 = run_logistic_model("target_30min_cls", "30 Minutes")
results_60, preds_60, probs_60, y_test_60 = run_logistic_model("target_60min_cls", "60 Minutes")

# Combine results
final_classification_results = pd.concat([results_10, results_30, results_60], ignore_index=True)

display(final_classification_results.round(3))

```

Logistic regression achieved high accuracy in general however class imbalance resulted in reduced precision for predicting empty stations. While the model reliably detects most empty cases (high recall), it has a high number og false positives.

### 3.2 Comparing Horizons

Comparing 10mins,30mins, and 60mins future prediction modelling

### 3.3 Discussion of Model Performance

Model performance was evaluated across multiple prediction horizons (10, 30, and 60 minutes).

At the 10-minute horizon, the persistence baseline performed almost as well as, and in terms of MAE slightly better than, the machine learning models, indicating strong short-term temporal stability in bike availability, where recent observations are highly predictive of near-future values.

However, as the prediction horizon increased, the performance of the baseline deteriorated substantially. At both 30 and 60 minutes, Linear Regression outperformed the baseline, suggesting that additional temporal and environmental features become more informative at longer forecasting horizons.

Linear Regression consistently outperformed Random Forest across all horizons. This suggests that the underlying relationships in the dataset are relatively linear, and that more complex models may introduce unnecessary variance or overfitting.

This is likely because bike availability changes quite gradually over time and follows fairly simple patterns. Linear Regression works well in this situation because it captures smooth, consistent trends, whereas Random Forest tends to break the data into lots of smaller segments, which makes it less suited to modelling these gradual changes.

In contrast, Random Forest showed noticeably weaker performance, particularly at longer horizons. This may be because tree-based models partition the data into discrete regions, which can make them less effective at modelling smooth, continuous changes. As a result, Random Forest may overfit to small fluctuations or noise rather than capturing the broader temporal patterns that drive bike availability.

In addition, Random Forest doesn’t naturally take the time-based structure of the data into account and can end up focusing too much on small fluctuations or noise, especially when many features are included. Its performance may also have been affected by limited tuning of its parameters. You could likely improve the Random Forest by using more trees and allowing them to grow a bit deeper so the model can pick up more patterns in the data. It would also help to tweak a few settings and remove less useful features, so it focuses on the important signals instead of getting distracted by noise.

Overall, these results suggest that simpler linear models are highly effective for this task, particularly when combined with well-engineered lag features, while more complex models such as Random Forest may not provide additional benefits without further tuning and refinement.

### 3.3 Testing 24hrs (longer horizon)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Create the prediction targets
df = df.copy()

df["target_10min"] = df.groupby("station_id")["num_bikes_available"].shift(-1)
df["target_30min"] = df.groupby("station_id")["num_bikes_available"].shift(-3)
df["target_60min"] = df.groupby("station_id")["num_bikes_available"].shift(-6)
df["target_24h"]   = df.groupby("station_id")["num_bikes_available"].shift(-144)

# 24h baseline feature created
df["bikes_avail_24h_ago"] = df.groupby("station_id")["num_bikes_available"].shift(144)

# Redefine features to allow for specifics
features = [
    "station_id", "lat", "lon", "capacity",
    "day", "hour", "minute",
    "day_of_week", "is_weekend",
    "avg_temp", "avg_humidity", "avg_barometric_pressure",
    "bikes_avail_10mins_ago", "bikes_avail_30mins_ago", "bikes_avail_1h_ago"
]


# Evaluation function redefined

def evaluate_horizon(df, target_col, baseline_col, features):
    required_cols = features + [target_col, baseline_col]
    temp_df = df.dropna(subset=required_cols).copy()

    X = temp_df[features]
    y = temp_df[target_col]
    baseline_series = temp_df[baseline_col]

    # time-based split
    split_idx = int(len(temp_df) * 0.8)
    X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
    y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]
    baseline_test = baseline_series.iloc[split_idx:]

    results = {}

    # Baseline
    y_pred_baseline = baseline_test
    results["Baseline"] = {
        "MAE": mean_absolute_error(y_test, y_pred_baseline),
        "RMSE": np.sqrt(mean_squared_error(y_test, y_pred_baseline)),
        "R2": r2_score(y_test, y_pred_baseline)
    }

    # Linear Regression
    lr = LinearRegression()
    lr.fit(X_train, y_train)
    preds_lr = lr.predict(X_test)

    results["Linear Regression"] = {
        "MAE": mean_absolute_error(y_test, preds_lr),
        "RMSE": np.sqrt(mean_squared_error(y_test, preds_lr)),
        "R2": r2_score(y_test, preds_lr)
    }

    # Random Forest
    rf = RandomForestRegressor(
        n_estimators=150,
        max_depth=None,
        min_samples_leaf=2,
        max_features="sqrt",
        n_jobs=-1,
        random_state=42
    )
    rf.fit(X_train, y_train)
    preds_rf = rf.predict(X_test)

    results["Random Forest"] = {
        "MAE": mean_absolute_error(y_test, preds_rf),
        "RMSE": np.sqrt(mean_squared_error(y_test, preds_rf)),
        "R2": r2_score(y_test, preds_rf)
    }

    return results


# Run all horizons (times)

all_results = {}

all_results["10 min"] = evaluate_horizon(
    df=df,
    target_col="target_10min",
    baseline_col="bikes_avail_10mins_ago",
    features=features
)

all_results["30 min"] = evaluate_horizon(
    df=df,
    target_col="target_30min",
    baseline_col="bikes_avail_30mins_ago",
    features=features
)

all_results["60 min"] = evaluate_horizon(
    df=df,
    target_col="target_60min",
    baseline_col="bikes_avail_1h_ago",
    features=features
)

# For 24h, use 24h baseline if available, otherwise fall back to 60min lag as previous
baseline_24h_col = "bikes_avail_24h_ago" if "bikes_avail_24h_ago" in df.columns else "bikes_avail_1h_ago"

all_results["24 h"] = evaluate_horizon(
    df=df,
    target_col="target_24h",
    baseline_col=baseline_24h_col,
    features=features
)

# Make results table

rows = []

for horizon, models in all_results.items():
    for model_name, metrics in models.items():
        rows.append({
            "Horizon": horizon,
            "Model": model_name,
            "MAE": metrics["MAE"],
            "RMSE": metrics["RMSE"],
            "R2": metrics["R2"]
        })

results_df = pd.DataFrame(rows)

# Make numeric version of horizon for plotting
horizon_map = {
    "10 min": 10,
    "30 min": 30,
    "60 min": 60,
    "24 h": 1440
}
results_df["Horizon_num"] = results_df["Horizon"].map(horizon_map)

print(results_df.sort_values(["Horizon_num", "Model"]))

# Visually plot reesults
import matplotlib.pyplot as plt

# Order horizons properly
order = ["10 min", "30 min", "60 min", "24 h"]
results_df["Horizon"] = pd.Categorical(results_df["Horizon"], categories=order, ordered=True)

plt.figure(figsize=(9,6))

colors = {
    "Baseline": "#97268a",
    "Linear Regression": "#5E949F",
    "Random Forest": "#925d02"
}

for model in results_df["Model"].unique():
    subset = results_df[results_df["Model"] == model].sort_values("Horizon")
    
    linewidth = 3 if model == "Linear Regression" else 2
    
    plt.plot(
        subset["Horizon"],
        subset["MAE"],
        marker="o",
        markersize=8,
        linewidth=linewidth,
        label=model,
        color=colors.get(model)
    )

# Only annotate key points (reduces clutter)
for _, row in results_df.iterrows():
    if row["Horizon"] in ["10 min", "24 h"]:  # only extremes to stop cluttered text
        plt.text(
            row["Horizon"],
            row["MAE"] + 0.15,
            f'{row["MAE"]:.2f}',
            ha="center",
            fontsize=9
        )

plt.title("Model Performance Across Prediction Horizons", fontsize=14, pad=15)
plt.xlabel("Prediction Horizon", fontsize=12)
plt.ylabel("Mean Absolute Error (MAE)", fontsize=12)

plt.grid(alpha=0.25)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.legend(frameon=False)
plt.tight_layout()
plt.show()
```

### 3.4 Results of Evaluation

The results show that prediction accuracy decreases as the time horizon increases for all models, which is expected as the task becomes more difficult. At the 10-minute horizon, the baseline performs as well as or slightly better than the machine learning models, indicating that short-term bike availability is highly stable and largely driven by recent observations.

However, as the prediction horizon extends to 30 and 60 minutes, the performance of the baseline declines, while Linear Regression consistently outperforms both the baseline and Random Forest. This suggests that while simple persistence works well in the short term, learned patterns become more important for longer-term predictions, with Linear Regression providing the most reliable performance at medium horizons.

At the 24-hour horizon, the baseline performs poorly, reflecting the breakdown of short-term persistence over longer time periods. In this case, Random Forest achieves the best performance, outperforming Linear Regression. This indicates that longer-term predictions are influenced by more complex and nonlinear patterns, such as daily cycles and interactions between features, which are better captured by tree-based models.

Overall, these results suggest that while simple persistence is sufficient for very short-term forecasting, machine learning models become necessary for longer horizons, with Linear Regression performing best at medium horizons and Random Forest showing advantages at longer-term prediction.

#### 4.3 Linear Regression for Short Term Model

```python

df = df.sort_values(["station_id", "date_time"]).reset_index(drop=True)

#Lag features
df["bikes_avail_10mins_ago"] = df.groupby("station_id")["num_bikes_available"].shift(1)
df["bikes_avail_30mins_ago"] = df.groupby("station_id")["num_bikes_available"].shift(3)
df["bikes_avail_1h_ago"] = df.groupby("station_id")["num_bikes_available"].shift(6)
df["bikes_avail_24h_ago"] = df.groupby("station_id")["num_bikes_available"].shift(144)

# targets
df["target_10min"] = df.groupby("station_id")["num_bikes_available"].shift(-1)
df["target_30min"] = df.groupby("station_id")["num_bikes_available"].shift(-3)
df["target_60min"] = df.groupby("station_id")["num_bikes_available"].shift(-6)
df["target_90min"] = df.groupby("station_id")["num_bikes_available"].shift(-9)
df["target_120min"] = df.groupby("station_id")["num_bikes_available"].shift(-18)

lag_cols = [
    "bikes_avail_10mins_ago",
    "bikes_avail_30mins_ago",
    "bikes_avail_1h_ago",
    "bikes_avail_24h_ago"
]

target_cols = [
    "target_10min",
    "target_30min",
    "target_60min",
    "target_90min",
    "target_120min"
]

df = df.dropna(subset=lag_cols + target_cols).reset_index(drop=True)

# Test/train split
df = df.sort_values("date_time").reset_index(drop=True)

cutoff_time = df["date_time"].quantile(0.8)
train_df = df[df["date_time"] <= cutoff_time].copy()
test_df = df[df["date_time"] > cutoff_time].copy()

#Features 
features = [
    "station_id",
    "capacity",
    "hour_sin",
    "hour_cos",
    "day_of_week",
    "is_weekend",
    "avg_temp",
    "avg_humidity",
    "avg_barometric_pressure",
    "bikes_avail_10mins_ago",
    "bikes_avail_30mins_ago",
    "bikes_avail_1h_ago",
    "bikes_avail_24h_ago"
]

categorical_features = ["station_id", "day_of_week"]

numeric_features = [
    "capacity",
    "hour_sin",
    "hour_cos",
    "is_weekend",
    "avg_temp",
    "avg_humidity",
    "avg_barometric_pressure",
    "bikes_avail_10mins_ago",
    "bikes_avail_30mins_ago",
    "bikes_avail_1h_ago",
    "bikes_avail_24h_ago"
]

# Raw feature matrices for pipeline
X_train = train_df[features].copy()
X_test = test_df[features].copy()

# Run model and save
def run_model(target_col, horizon_label, pickle_file=None):
    y_train = train_df[target_col].copy()
    y_test = test_df[target_col].copy()

    preprocessor = ColumnTransformer(
        transformers=[
            ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), categorical_features),
            ("num", "passthrough", numeric_features)
        ]
    )

    pipeline = Pipeline([
        ("preprocessor", preprocessor),
        ("model", LinearRegression())
    ])

    pipeline.fit(X_train, y_train)
    preds = pipeline.predict(X_test)

    if pickle_file is not None:
        with open(pickle_file, "wb") as f:
            pickle.dump(pipeline, f)

    results = pd.DataFrame([{
        "Horizon": horizon_label,
        "Model": "Linear Regression",
        "MAE": mean_absolute_error(y_test, preds),
        "RMSE": np.sqrt(mean_squared_error(y_test, preds)),
        "R2": r2_score(y_test, preds)
    }])

    return results

# Model for each horizon
results_10 = run_model("target_10min", "10 Minutes", "lr_10min.pkl")
results_30 = run_model("target_30min", "30 Minutes", "lr_30min.pkl")
results_60 = run_model("target_60min", "60 Minutes", "lr_60min.pkl")
results_90 = run_model("target_90min", "90 Minutes", "lr_90min.pkl")
results_120 = run_model("target_120min", "120 Minutes", "lr_120min.pkl")

# results
final_results = pd.concat(
    [results_10, results_30, results_60, results_90, results_120],
    ignore_index=True
)

final_results = final_results.sort_values(by=["Horizon", "MAE"]).reset_index(drop=True)

display(final_results.round(3))
```

## 1. Data Preparation/Cleaning

### 1.1 Introduction - import libraries and read in CSV file

```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def load_data(file):
    # Read CSV file into a pandas dataframe
    # Print first few rows

    try:
        # If the csv file has any special symbols, it helps to specify the file encoding.
        df = pd.read_csv(file, 
            encoding="utf-8",
            delimiter=",",
            skipinitialspace=True
        )
        
        print(f"✓ Successfully loaded dataset with {df.shape[0]} rows and {df.shape[1]} columns")

        return df
    
    except FileNotFoundError:
        print("ERROR: The file '" + file + "' was not found.")
        print("Please ensure the file is in the current working directory.")
    except Exception as e:
        print(f"ERROR: An unexpected error occurred: {e}")

df_test = load_data("final_merged_data.csv")
df_raw = df_test.copy()
```

### 1.2 Data Validation


```python
def inspect_data(df):
    # Data validation checks

    print("Dataset Validation Checks:")
    print("="*60)

    # Check 1: Dataset is not empty
    if df.empty:
        print("⚠ WARNING: Dataset is empty!")
    else:
        print(f"✓ Dataset contains {df.shape[0]:,} rows")

    # Check 2: Check for duplicate rows
    duplicate_count = df.duplicated().sum()
    if duplicate_count > 0:
        print(f"⚠ WARNING: Found {duplicate_count} duplicate rows")
    else:
        print("✓ No duplicate rows found")

    # Check 3: Check if all columns are unnamed
    unnamed_cols = [col for col in df.columns if 'Unnamed' in str(col)]
    if unnamed_cols:
        print(f"⚠ WARNING: Found {len(unnamed_cols)} unnamed columns: {unnamed_cols}")
    else:
        print("✓ All columns have names")

    # Check 4: Check for missing data
    missing_count = df.isna().sum().sum()
    if missing_count > 0:
        print(f"⚠ WARNING: Found {missing_count} missing values")
    else:
        print("✓ No missing values found")

    # Check 5: Check for constant columns (cardinality of 1)
    constant_cols = df.columns[df.nunique() == 1].tolist()

    if constant_cols:
        print(f"\n⚠ WARNING: Constant columns found: {constant_cols}")
    else:
        print(f"✓ No constant columns found — all {df.shape[1]} features contain variation")

    # Check 5: Basic column overview
    print(f"\n✓ Dataset shape: {df.shape}")
    print(f"✓ Number of rows (instances): {df.shape[0]}")
    print(f"✓ Number of columns (features): {df.shape[1]}")

inspect_data(df_test)
```

### 1.3 Data Base Cleaning

```python
def base_clean(df):
    df = df.copy()
    df.columns = df.columns.str.strip().str.lower()

    # Parse last_reported into a datetime column
    df["date_time"] = pd.to_datetime(df["last_reported"], errors="coerce")

    # Drop duplicates and missing key values (not present in our dataset but included in case of use with future data)
    df = df.dropna(subset=["date_time", "station_id", "num_bikes_available", "capacity"])
    df = df.drop_duplicates()

    # Drop constant columns
    constant_cols = [c for c in df.columns if df[c].nunique() == 1]
    df = df.drop(columns=constant_cols)

    # Drop invalid columns
    df = df[(df['num_bikes_available'] >= 0) & 
        (df['num_bikes_available'] <= df['capacity'])]

    #Change station ID feature to categorical
    df["station_id"] = df["station_id"].astype("category")
    
    return df

df_test = base_clean(df_test)
```

### 1.4 Feature preparation

Cleaning decisions made:
- Parse last_reported into datetime column (possibly to be used to derive lag features during modelling)
- Drop rows with missing values or invalid values (e.g. num_bikes_available > capacity)
- Drop duplicate rows
- Drop constant columns
- Drop num_docks_available to prevent data leakage
- Drop redundant features (e.g. std columns)
- Drop categorical features not relevent for modelling (e.g. name, address, last reported)
- Drop soil/earth/grass features as not thought to be relevent for modelling

```python
def drop_columns(df):

    #Drop available docks feature - avoid data leakage
    df = df.drop(columns=["num_docks_available"], errors="ignore")
    
    # Drop quality indicator columns (should already be removed when removing constants)
    quality_indicator_cols = [c for c in df.columns if "quality_indicator" in c.lower()]
    df = df.drop(columns=quality_indicator_cols)

    # Drop earth/soil feature columns  - not statistically relevant 
    earth_cols = [c for c in df.columns if any(x in c.lower() for x in ["soil", "earth", "grass"])]
    df = df.drop(columns=earth_cols)

    # Drop std columns - redundancy
    std_cols = [c for c in df.columns if "std_deviation" in c.lower()]
    df = df.drop(columns=std_cols)

    # Drop other irrelevant columns
    df = df.drop(columns=["name", "address", "last_reported"], errors="ignore")

    return df
```

### 1.5 Derived features

```python
# Week day/ weekend indicators
def add_time_features(df):

    df["day_of_week"] = df["date_time"].dt.dayofweek
    df["is_weekend"] = (df["day_of_week"] >= 5).astype(int)

    return df

# Average environmental features
def add_average_features(df):

    #Average temp
    df["avg_temp"] = (df["max_air_temperature_celsius"] + df["min_air_temperature_celsius"]) / 2

    #Average humidity
    df["avg_humidity"] = (df["max_relative_humidity_percent"] + df["min_relative_humidity_percent"]) / 2

    #Average pressure
    df["avg_barometric_pressure"] = (df["max_barometric_pressure_hpa"] + df["min_barometric_pressure_hpa"]) / 2

    # Drop min/max/std features
    df = df.drop(columns=[
        "max_air_temperature_celsius",
        "min_air_temperature_celsius",
        "max_relative_humidity_percent",
        "min_relative_humidity_percent",
        "max_barometric_pressure_hpa", 
        "min_barometric_pressure_hpa"
        ])
    return df

# Add lag features
def add_lag_features(df):
    df = df.copy()
    df = df.sort_values(["station_id", "date_time"])

    # Available bikes 10 mins ago
    df["bikes_avail_10mins_ago"] = df.groupby("station_id")["num_bikes_available"].shift(1)

    # Available bikes 30 mins ago
    df["bikes_avail_30mins_ago"] = df.groupby("station_id")["num_bikes_available"].shift(3)

    # Available bikes 1 hour ago
    df["bikes_avail_1h_ago"] = df.groupby("station_id")["num_bikes_available"].shift(6)

    # Available bikes same time yesterday
    df["bikes_avail_24h_ago"] = df.groupby("station_id")["num_bikes_available"].shift(144)

    # Avilable bikes same time 1w ago
    df["bikes_avail_1w_ago"] = df.groupby("station_id")["num_bikes_available"].shift(1008)

    # List of lag columns
    lag_cols = [
        "bikes_avail_10mins_ago",
        "bikes_avail_30mins_ago",
        "bikes_avail_1h_ago",
        "bikes_avail_24h_ago",
        "bikes_avail_1w_ago"
    ]

    # Add indicators + fill missing values
    for col in lag_cols:
        df[f"has_{col}"] = df[col].notna().astype(int)
        df[col] = df[col].fillna(0)

    return df

# Add ratio feature
def add_bike_ratio(df):
    df["bike_ratio"] = df["num_bikes_available"] / df["capacity"]
    return df
```

### 1.6 Apply changes and save new file

```python
df_cleaned = (
    load_data("final_merged_data.csv")
    .pipe(base_clean)
    .pipe(drop_columns)
    .pipe(add_time_features)
    .pipe(add_average_features)
    .pipe(add_lag_features)
    .pipe(add_bike_ratio)
)
```

```python
print("="*60)
print(f"Before cleaning: {df_raw.shape[0]:,} rows, {df_raw.shape[1]} columns")

rows_removed = df_raw.shape[0] - df_cleaned.shape[0]
cols_removed = df_raw.shape[1] - df_cleaned.shape[1]

print("="*60)
print(f"Rows removed: {rows_removed}")
print(f"Columns removed: {cols_removed}")

print("="*60)
print(f"After cleaning:  {df_cleaned.shape[0]:,} rows, {df_cleaned.shape[1]} columns")

print("="*60)
print(df_cleaned.info())
```

```python
#Save new cleaned file 
df_cleaned.to_csv("CleanedData.csv", index=False)
```

#### 2.1 Linear Regression for Long Term Prediction

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import joblib
import pickle

df["date_time"] = pd.to_datetime(df["date_time"])
df = df.sort_values('date_time')

# Drop missing values (if any remain)
df.dropna(inplace=True)

# Define features and target
features = [
    "station_id",
    "capacity",
    "hour",
    "minute",
    "day_of_week",
    "is_weekend",
    "avg_temp",
    "avg_humidity",
    "avg_barometric_pressure"
]

target = "num_bikes_available"

# Define train test split - time based
split_index = int(len(df) * 0.8)

train_df = df.iloc[:split_index].copy()
test_df = df.iloc[split_index:].copy()

print("Train date range:", train_df["date_time"].min(), "to", train_df["date_time"].max())
print("Test date range: ", test_df["date_time"].min(), "to", test_df["date_time"].max())

group_cols = ["station_id", "day_of_week", "hour"]

# Define historic averages from train data
station_time_avg = (
    train_df.groupby(group_cols)[target]
    .mean()
    .rename("avg_bikes_station_time")
    .reset_index()
)

station_avg = (
    train_df.groupby("station_id")[target]
    .mean()
    .rename("station_avg")
    .reset_index()
)

global_avg = train_df[target].mean()

# merge train-derived averages into both train and test
train_df = train_df.merge(station_time_avg, on=group_cols, how="left")
test_df = test_df.merge(station_time_avg, on=group_cols, how="left")

train_df = train_df.merge(station_avg, on="station_id", how="left")
test_df = test_df.merge(station_avg, on="station_id", how="left")

# fill missing values
for d in [train_df, test_df]:
    d["avg_bikes_station_time"] = d["avg_bikes_station_time"].fillna(d["station_avg"])
    d["avg_bikes_station_time"] = d["avg_bikes_station_time"].fillna(global_avg)


# Define final features
final_features = features + ["avg_bikes_station_time"]

X_train = train_df[final_features].copy()
X_test = test_df[final_features].copy()
y_train = train_df[target].copy()
y_test = test_df[target].copy()

# One-hot encode categoricals
X_train = pd.get_dummies(X_train, columns=["station_id", "day_of_week"], drop_first=True)
X_test = pd.get_dummies(X_test, columns=["station_id", "day_of_week"], drop_first=True)

# Ensure test has same columns as train
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# Train model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predictions
y_pred = lr_model.predict(X_test)

# Evaluation
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae:.4f}")
print(f"R² Score: {r2:.4f}")

# Display model coefficients
print("\nModel Coefficients:")
for feature, coef in zip(X_train.columns, lr_model.coef_):
    print(f"{feature}: {coef}")
print(f"Intercept: {lr_model.intercept_}")

```

The baseline model, which used bike availability from 10 minutes earlier as the prediction, significantly outperformed all machine learning models, achieving a much lower MAE (1.26) and higher R² (0.93). Random Forest performed the next best, but still signifianctly worse with an MAE (3.657) and RMSE (4.893) and the highest R² score (0.749). This indicates strong predictive accuracy and the ability to capture non-linear relationships in the data. The Decision Tree model also performed well, but with slightly higher error metrics. In contrast, both Linear Regression and Ridge Regression produced identical results, with higher MAE and RMSE and lower R². This suggests that linear models are less useful for this problem. Overall, the results demonstrate that ensemble methods such as Random Forest are better suited to modelling the complex patterns in bike availability. However, it is important to note that the inclusion of engineered features such as average bike availability may introduce some data leakage, meaning the reported performance could be slightly optimistic compared to real-world deployment.

#### 2.6 Time Series for Short Term Prediction

```python
import sys
import importlib

mods_to_remove = [
    m for m in list(sys.modules)
    if m.startswith("sktime")
]

for m in mods_to_remove:
    del sys.modules[m]

importlib.invalidate_caches()

from sktime.forecasting.ets import AutoETS
print("AutoETS imported successfully")
```

```python
from sktime.forecasting.base import ForecastingHorizon
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.forecasting.ets import AutoETS
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error

# Choose one station
station_id = 42
station_df = df[df["station_id"] == station_id].copy()

# Sort by time
station_df = station_df.sort_values("date_time")

# Keep only the target series
y = station_df.set_index("date_time")["num_bikes_available"]

# Force regular 10-minute frequency
y = y.asfreq("10min")

# Fill any missing timestamps if needed
y = y.interpolate()

# Convert to PeriodIndex for sktime
y.index = pd.PeriodIndex(y.index, freq="10min")

# Split data into train and test sets
y_train, y_test = temporal_train_test_split(y, test_size=144)

# Daily seasonality for 10-minute data = 144 periods per day
forecaster = AutoETS(auto=True, sp=144, n_jobs=-1)
forecaster.fit(y_train)

# Forecast horizon
fh = ForecastingHorizon(y_test.index, is_relative=False)

# Generate forecasts
y_pred = forecaster.predict(fh)

# Calculate evaluation metric
mape = mean_absolute_percentage_error(y_test, y_pred)

# Convert indices for plotting
y_train.index = y_train.index.to_timestamp()
y_test.index = y_test.index.to_timestamp()
y_pred.index = y_pred.index.to_timestamp()

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(y_train.index, y_train, label="Training Data")
plt.plot(y_test.index, y_test, label="Actual Values")
plt.plot(y_pred.index, y_pred, label="Forecast")
plt.title(f"Bike Availability Forecast for Station {station_id} (MAPE: {mape:.2%})")
plt.xlabel("Date")
plt.ylabel("Available Bikes")
plt.legend()
plt.grid(True)
plt.show()
```
